```python
docs_ = []

cities = {
    "Nairobi": "KE",
    "Lagos": "NG",
    "Johannesburg": "ZA",
    "Accra": "GH"
}

for city, country in cities.items():
    params = {
        "q": f"{city},{country}",
        "appid": API_KEY,
        "units": "metric"
    }
    print(f"Fetching weather data for {city}...")
    resp = requests.get(BASE_URL, params=params)
    if resp.status_code == 200:
        data = resp.json()
        pprint(data)
        docs_.append(data)
    else:
        print(f"Error Fetching data for {city}: {resp.status_code} - {resp.text}")
```
